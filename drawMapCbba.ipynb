```python
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
import pandas as pd

dep = pd.read_csv('Departamentos_Bolivia_2015.csv')
#dep.dtypes
```

```python
dep
```

```python
strLen = len(dep['the_geom'][0]) - 3
#dep['the_geom'][0][16:50]
cocha = dep['the_geom'][0][16:strLen]
```

```python
sinComa = cocha.split(',')
cochaPoly = {}
for i in range(len(sinComa)):
    cochaPoly[i] = [float(sinComa[i].split()[0]), float(sinComa[i].split()[1])]
```

```python
dfCocha = pd.DataFrame(cochaPoly, index = ['longitude', 'latitude']).T
BBox = (dfCocha.longitude.min(), dfCocha.longitude.max(), dfCocha.latitude.min(), dfCocha.latitude.max())
```

```python
fig, ax = plt.subplots(figsize = (8,7))

ax.fill(dfCocha.longitude, dfCocha.latitude, facecolor='none', edgecolor='blue', linewidth=1)

ax.set_xlim(BBox[0],BBox[1])
ax.set_ylim(BBox[2],BBox[3])
```
